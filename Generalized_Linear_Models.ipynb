## Sufficient Statistics
A *statistic* $T$ is a function $T = r(X_1, X_2, \dots, X_n)$ of the random sample $X_1, X_2, \dots, X_n$. The mean, median and variance of the sample are all examples of statistics.
$$
\bar{X}_{n} = \frac{1}{n}\sum_{i=1}^{n}X_i \quad \text{(the sample mean)} \\
\text{var}(X_{n}) = s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X}_{n})^2 \quad \text{(the sample variance)} \\
$$
A statistic is used to estimate the unknown parameter $\theta$ of a population from a random sample drawn from that population. For example, the sample mean $\bar{X}_{n}$ is an estimator of the population mean $\mu$. The sample variance $\text{var}(X_{n})$ is an estimator of the population variance $\sigma^2$.

A statistic is an **estimator** of a population if $T$ is close to $\theta$ with high probability as the sample size $n$ increases. Formally, we say that $T$ is a **consistent estimator** of $\theta$ if for every $\epsilon > 0$, we have:
$$
\lim_{n \to \infty} P(|T - \theta| > \epsilon) = 0
$$

Out of all statistics, we are interested in those that are good estimators. To find such statistics, we can use the concept of **sufficient statistics**. When we say a statistic $T$ is *sufficient*, it means that a statistician who knows $T$ can do just as good a job of estimating $\theta$ as a statistician who knows the entire sample. Formally, a statistic $T$ is sufficient for $\theta$ if the conditional distribution of the sample given $T$ does not depend on $\theta$. In other words, we have:
$$
f_{X}(X_1, X_2, \dots, X_n \mid T) = f_{X}(X_1, X_2, \dots, X_n \mid T, \theta)
$$

**Factorization Theorem:** Let $X_1, X_2, \dots, X_n$ be a random sample from a distribution with probability density function (pdf) $f(x; \theta)$. A statistic $T = r(X_1, X_2, \dots, X_n)$ is sufficient for $\theta$ if and only if it can be factored into two non-negative functions:
$$
f(X; \theta) = g(T(X); \theta) h(X)
$$

For example, for a normal distribution with unknown mean $\mu$ and known variance $\sigma^2$, the PDF is:
$$
f(x; \mu) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

Assuming $X_{i}$ are i.i.d. samples from this distribution, the joint PDF is:
$$
\begin{align*}
f(X; \mu) &= \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(X_i - \mu)^2}{2\sigma^2}\right) \\ 
&= \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (X_i - \mu)^2\right) \\
&= \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (X_i^2 - 2X_i\mu + \mu^2)\right) \\
&= \underbrace{\left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} X_i^2\right)}_{h(X)} \underbrace{\exp\left(\frac{\mu}{\sigma^2} \underbrace{\sum_{i=1}^nX_i}_{T(X)} - \frac{n\mu^2}{2\sigma^2}\right)}_{g(T(X); \mu)} \\
\end{align*}
$$

Thus, by the factorization theorem, $T(X) = \sum_{i=1}^n X_i$ is a sufficient statistic for $\mu$. If we think about it, this is actually the same as saying that the sample mean $\bar{X}_{n} = \frac{1}{n}\sum_{i=1}^n X_i$ is a sufficient statistic for $\mu$, since $\bar{X}_{n}$ is just a scaled version of $T(X)$.

## Exponential Family Distributions
We say that a probability distribution belongs to the exponential family if it can be expressed in the following form:
$$
p(y; \eta) = b(y) \exp(\eta^{T}T(y) - a(\eta))
$$
where $\eta$ is the **natural parameter**, $T(y)$ is the **sufficient statistic**, $a(\eta)$ is the **log-partition** function, and $b(y)$ is the **base measure**.
