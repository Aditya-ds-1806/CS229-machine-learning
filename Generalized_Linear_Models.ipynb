## Sufficient Statistics
A *statistic* $T$ is a function $T = r(X_1, X_2, \dots, X_n)$ of the random sample $X_1, X_2, \dots, X_n$. The mean, median and variance of the sample are all examples of statistics.
$$
\bar{X}_{n} = \frac{1}{n}\sum_{i=1}^{n}X_i \quad \text{(the sample mean)} \\
\text{var}(X_{n}) = s^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \bar{X}_{n})^2 \quad \text{(the sample variance)} \\
$$
A statistic is used to estimate the unknown parameter $\theta$ of a population from a random sample drawn from that population. For example, the sample mean $\bar{X}_{n}$ is an estimator of the population mean $\mu$. The sample variance $\text{var}(X_{n})$ is an estimator of the population variance $\sigma^2$.

A statistic is an **estimator** of a population if $T$ is close to $\theta$ with high probability as the sample size $n$ increases. Formally, we say that $T$ is a **consistent estimator** of $\theta$ if for every $\epsilon > 0$, we have:
$$
\lim_{n \to \infty} P(|T - \theta| > \epsilon) = 0
$$

Out of all statistics, we are interested in those that are good estimators. To find such statistics, we can use the concept of **sufficient statistics**. When we say a statistic $T$ is *sufficient*, it means that a statistician who knows $T$ can do just as good a job of estimating $\theta$ as a statistician who knows the entire sample. Formally, a statistic $T$ is sufficient for $\theta$ if the conditional distribution of the sample given $T$ does not depend on $\theta$. In other words, we have:
$$
f_{X}(X_1, X_2, \dots, X_n \mid T) = f_{X}(X_1, X_2, \dots, X_n \mid T, \theta)
$$

**Factorization Theorem:** Let $X_1, X_2, \dots, X_n$ be a random sample from a distribution with probability density function (pdf) $f(x; \theta)$. A statistic $T = r(X_1, X_2, \dots, X_n)$ is sufficient for $\theta$ if and only if the joint PDF can be factored into two non-negative functions:
$$
f(X; \theta) = g(T(X); \theta) h(X)
$$

For example, for a normal distribution with unknown mean $\mu$ and known variance $\sigma^2$, the PDF is:
$$
f(x; \mu) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

Assuming $X_{i}$ are i.i.d. samples from this distribution, the joint PDF is:
$$
\begin{align*}
f(X; \mu) &= \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(X_i - \mu)^2}{2\sigma^2}\right) \\ 
&= \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (X_i - \mu)^2\right) \\
&= \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (X_i^2 - 2X_i\mu + \mu^2)\right) \\
&= \underbrace{\left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} X_i^2\right)}_{h(X)} \underbrace{\exp\left(\frac{\mu}{\sigma^2} \underbrace{\sum_{i=1}^nX_i}_{T(X)} - \frac{n\mu^2}{2\sigma^2}\right)}_{g(T(X); \mu)} \\
\end{align*}
$$

Thus, by the factorization theorem, $T(X) = \sum_{i=1}^n X_i$ is a sufficient statistic for $\mu$. If we think about it, this is actually the same as saying that the sample mean $\bar{X}_{n} = \frac{1}{n}\sum_{i=1}^n X_i$ is a sufficient statistic for $\mu$, since $\bar{X}_{n}$ is just a scaled version of $T(X)$.

## Exponential Family Distributions
We say that a probability distribution belongs to the exponential family if it can be expressed in the following form:
$$
p(y; \eta) = b(y) \exp(\eta^{T}T(y) - a(\eta))
$$
where $\eta$ is the **natural parameter**, $T(y)$ is the **sufficient statistic**, $a(\eta)$ is the **log-partition** function, and $b(y)$ is the **base measure**. The log-partition function is defined as:
$$
a(\eta) = \log \int b(y) \exp(\eta^{T}T(y)) dy
$$

### Bernoulli Distribution belongs to the Exponential Family
The probability mass function (PMF) of a Bernoulli distribution with parameter $\phi$ is given by:
$$
p(y; \phi) = \phi^y (1-\phi)^{1-y} \quad \text{for } y \in \{0, 1\}
$$
where $\phi$ is the probability of success (i.e., $P(Y=1) = \phi$). We can show that the Bernoulli distribution belongs to the exponential family:

$$
\begin{align*}
p(y; \phi) &= \phi^y (1-\phi)^{1-y} \\
&= \exp\left(\log(\phi^y (1-\phi)^{1-y})\right) \\
&= \exp\left(y \log(\phi) + (1-y) \log(1-\phi)\right) \\
&= \exp\left(\underbrace{y}_{T(y)} \underbrace{\log\left(\frac{\phi}{1-\phi}\right)}_{\eta} + \underbrace{\log(1-\phi)}_{-a(\phi)}\right) \\

\implies& \eta = \log\left(\frac{\phi}{1-\phi}\right) \\
& \phi = \frac{1}{1 + e^{-\eta}} \\
& T(y) = y \\
& a(\eta) = -\log(1-\phi) = \log(1 + e^{\eta}) \\
& b(y) = 1\quad\blacksquare \\
\end{align*}
$$

We will now show the $T(y) = y$ is indeed a sufficient statistic for the Bernoulli distribution. The joint PMF of $n$ i.i.d. samples from a Bernoulli distribution is given by:
$$
\begin{align*}
p(Y; \phi) &= \prod_{i=1}^{n} \phi^{Y_i} (1-\phi)^{1-Y_i} \\
&= \exp\left(\sum_{i=1}^{n} Y_i \log(\phi) + (1-Y_i) \log(1-\phi)\right) \\
&= \exp\left(\log(\phi) \sum_{i=1}^{n} Y_i + \log(1-\phi) \sum_{i=1}^{n} (1-Y_i)\right) \\
&= \exp\left(\log(\phi) \sum_{i=1}^{n} Y_i + \log(1-\phi) \left(n - \sum_{i=1}^{n} Y_i\right)\right) \\
&= \exp\left(\log\left(\frac{\phi}{1-\phi}\right) \sum_{i=1}^{n} Y_i + n \log(1-\phi)\right) \\
&= \underbrace{\exp\left(n \log(1-\phi)\right)}_{h(Y)} \exp\left(\underbrace{\sum_{i=1}^{n} Y_i}_{T(Y)} \underbrace{\log\left(\frac{\phi}{1-\phi}\right)}_{\eta}\right) \\
\end{align*}
$$

When $Y$ is a scalar, $T(Y)$ becomes $T(y) = y$. $\blacksquare$

We will now show that $a(\eta)$ is indeed the log-partition function for the Bernoulli distribution. Since $y$ can only take values 0 and 1, we have:
$$
\begin{align*}
a(\eta) &= \log \int b(y) \exp(\eta^{T}T(y)) dy \\
&= \log \sum_{y=0}^{1} b(y) \exp(\eta^{T}T(y)) \\
&= \log \left( b(0) \exp(\eta^{T}T(0)) + b(1) \exp(\eta^{T}T(1)) \right) \\
&= \log \left( 1 \cdot \exp(0) + 1 \cdot \exp(\eta) \right) \\
&= \log(1 + e^{\eta}) \blacksquare \\
\end{align*}
$$

### Gaussian Distribution with known variance belongs to the Exponential Family
The probability density function (PDF) of a Gaussian distribution with mean $\mu$ and known variance $\sigma^2$ is given by:
$$
p(y; \mu) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y - \mu)^2}{2\sigma^2}\right)
$$
We can show that the Gaussian distribution with known variance (assume $\sigma = 1$) belongs to the exponential family:
$$
\begin{align*}
p(y; \mu) &= \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(y - \mu)^2}{2}\right) \\
&= \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{y^2 - 2y\mu + \mu^2}{2}\right) \\
&= \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{y^2}{2} + y\mu - \frac{\mu^2}{2}\right) \\
&= \underbrace{\frac{1}{\sqrt{2\pi}} \exp\left(-\frac{y^2}{2}\right)}_{b(y)} \exp\left(\underbrace{y}_{T(y)}\underbrace{\mu}_{\eta} - \underbrace{\frac{\mu^2}{2}}_{a(\mu)}\right) \\

\implies& \eta = \mu \\
& T(y) = y \\
& a(\eta) = \frac{\eta^2}{2} \\
& b(y) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{y^2}{2}\right) \\
\end{align*}
$$

We will now show that $T(y) = y$ is indeed a sufficient statistic for the Gaussian distribution with known variance. The joint PDF of $n$ i.i.d. samples from a Gaussian distribution with known variance is given by:
$$
\begin{align*}
f(Y; \mu) &= \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(Y_i - \mu)^2}{2}\right) \\
&= \left(\frac{1}{\sqrt{2\pi}}\right)^n \exp\left(-\frac{1}{2} \sum_{i=1}^{n} (Y_i - \mu)^2\right) \\
&= \left(\frac{1}{\sqrt{2\pi}}\right)^n \exp\left(-\frac{1}{2} \sum_{i=1}^{n} (Y_i^2 - 2Y_i\mu + \mu^2)\right) \\
&= \underbrace{\left(\frac{1}{\sqrt{2\pi}}\right)^n \exp\left(-\frac{1}{2} \sum_{i=1}^{n} Y_i^2\right)}_{h(Y)} \exp\left(\underbrace{\sum_{i=1}^{n} Y_i}_{T(Y)} \underbrace{\mu}_{\eta} - \underbrace{\frac{n\mu^2}{2}}_{a(\mu)}\right) \\
\end{align*}
$$
When $Y$ is a scalar, $T(Y)$ becomes $T(y) = y$. $\blacksquare$

We will now show that $a(\eta)$ is indeed the log-partition function for the Gaussian distribution with known variance. We have:
$$
\begin{align*}
a(\eta) &= \log \int b(y) \exp(\eta^{T}T(y)) dy \\
&= \log \int \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{y^2}{2} + y \eta\right) dy \\
&= \log \int \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(y - \eta)^2}{2} + \frac{\eta^2}{2}\right) dy \\
&= \log \left( \exp\left(\frac{\eta^2}{2}\right) \int \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{(y - \eta)^2}{2}\right) dy \right) \\
&= \log \left( \exp\left(\frac{\eta^2}{2}\right) \cdot 1 \right) \\
&= \frac{\eta^2}{2} \blacksquare \\
\end{align*}
$$

### Properties of Exponential Family Distributions
1. The **base measure is non-negative**, i.e., $b(y) \ge 0$ for all $y$. Reason: If $b(y) < 0$ for some $y$, then the probability density function (PDF) would be negative for that value of $y$, which is not possible since probabilities cannot be negative.
2. The log likelihood of the exponential family of distributions is **concave** in the natural parameter $\eta$. This can be shown by taking the second derivative of the log likelihood with respect to $\eta$ and showing that it is negative.
    $$
        \begin{align*}
        \ell(\eta) &= \log p(y; \eta) = \log b(y) + \eta^{T}T(y) - a(\eta) \\
        \frac{\partial \ell(\eta)}{\partial \eta} &= T(y) - \frac{\partial a(\eta)}{\partial \eta} \\
        \frac{\partial^2 \ell(\eta)}{\partial \eta^2} &= -\frac{\partial^2 a(\eta)}{\partial \eta^2} \\
        &= -\frac{\partial^2}{\partial \eta^2}\int b(y) \exp(\eta^{T}T(y)) dy \\
        &= -\int \underbrace{b(y)}_{\ge 0} \underbrace{T(y) T(y)^{T}}_{\ge 0} \underbrace{\exp(\eta^{T}T(y))}_{\ge 0} dy \\
        &\preceq 0 \blacksquare \\
        \end{align*}
    $$
    This is important because it means that the maximum likelihood estimator (MLE) for $\eta$ can be found by setting the first derivative of the log likelihood to zero and solving for $\eta$. Since the log likelihood is concave, this will give us a single unique global maximum.
3. The expectation of the exponential family of distributions can be expressed in terms of the log-partition function $a(\eta)$ as follows:
    $$
        \begin{align*}
        \mathbb{E}[y; \eta] &= \int y p(y; \eta) dy \\
        &= \int y b(y) \exp(\eta^{T}y - a(\eta)) dy \\
        &= \exp(-a(\eta)) \int y b(y) \exp(\eta^{T}y) dy \\
        &= \exp(-a(\eta)) \frac{\partial}{\partial \eta} \int b(y) \exp(\eta^{T}y) dy \quad \left(\because \frac{\partial}{\partial \eta} \exp(\eta^{T}y) = y \exp(\eta^{T}y) \right) \\
        &= \exp(-a(\eta)) \frac{\partial}{\partial \eta} \exp(a(\eta)) \quad \left(\because \exp(a(\eta)) = \int b(y) \exp(\eta^{T}y) dy \right) \\
        &= \exp(-a(\eta)) \exp(a(\eta)) \frac{\partial a(\eta)}{\partial \eta} \\
        &= \frac{\partial a(\eta)}{\partial \eta} \blacksquare
        \end{align*}
    $$

4. Variance of the exponential family of distributions can also be expressed in terms of the log-partition function $a(\eta)$ as follows:
    $$
        \begin{align*}
        \text{Var}[y; \eta] &= \mathbb{E}[y^2; \eta] - (\mathbb{E}[y; \eta])^2 \\
        &= \int y^2 p(y; \eta) dy - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \\
        &= \int y^2 b(y) \exp(\eta^{T}y - a(\eta)) dy - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \\
        &= \exp(-a(\eta)) \int y^2 b(y) \exp(\eta^{T}y) dy - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \\
        &= \exp(-a(\eta)) \frac{\partial^2}{\partial \eta^2} \int b(y) \exp(\eta^{T}y) dy - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \quad \left(\because \frac{\partial^2}{\partial \eta^2} \exp(\eta^{T}y) = y^2 \exp(\eta^{T}y) \right) \\
        &= \exp(-a(\eta)) \frac{\partial^2}{\partial \eta^2} \exp(a(\eta)) - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \quad \left(\because \exp(a(\eta)) = \int b(y) \exp(\eta^{T}y) dy \right) \\
        &= \exp(-a(\eta)) \frac{\partial}{\partial \eta} \left(\exp(a(\eta)) \frac{\partial a(\eta)}{\partial \eta}\right) - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2  \\
        &= \exp(-a(\eta)) \left( \exp(a(\eta)) \frac{\partial^2 a(\eta)}{\partial \eta^2} + \exp(a(\eta)) \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \right) - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2  \\
        &= \exp(-a(\eta)) \exp(a(\eta)) \left( \frac{\partial^2 a(\eta)}{\partial \eta^2} + \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 \right) - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2  \\
        &= \frac{\partial^2 a(\eta)}{\partial \eta^2} + \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2 - \left(\frac{\partial a(\eta)}{\partial \eta}\right)^2  \\ 
        &= \frac{\partial^2 a(\eta)}{\partial \eta^2} \blacksquare  \\ 
        \end{align*}
    $$

## Generalized Linear Models (GLMs)
It turns out that linear regression and logistic regression are just special cases of a more general class of models called **generalized linear models (GLMs)**. GLMs are a flexible generalization of linear regression that allows for the **response variable** $y$ (i.e., the target variable) to have a distribution other than the normal distribution. To construct GLMs, we make the following assumptions:
1. $p(y \mid x; \theta) \sim \text{ExponentialFamily}(\eta)$, i.e., the distribution of the response variable $y$ given the feature vector $x$ and parameters $\theta$ belongs to the exponential family of distributions.
2. The hypothesis function $h_{\theta}(x)$ is defined as the expected value of $y$ given $x$ and $\theta$, i.e., $h_{\theta}(x) = \mathbb{E}[y \mid x; \theta]$.
3. The natural parameter $\eta$ is a linear function of the input features $x$, i.e., $\eta = \theta^{T}x$, where $\theta$ is the parameter vector and $x$ is the feature vector.

We will now show that the choice of the hypothesis function we made in linear regression and logistic regression is actually a consequence of the assumptions we made to construct GLMs.

### Linear Regression is a GLM
To show that linear regression is a special case of GLMs, we model the response variable $y$ as a Gaussian distribution with mean $\mu$ and variance $\sigma^2$:
$$
\begin{align*}
p(y \mid x; \theta) &\sim \mathcal{N}(\mu, \sigma^2) \\
h_{\theta}(x) &= \mathbb{E}[y \mid x; \theta] \\
&= \mu \\
&= \eta \quad\left(\because \eta = \mu \text{ for Gaussian}\right) \\
&= \theta^{T}x \blacksquare \\
\end{align*}
$$

### Logistic Regression is a GLM
To show that logistic regression is a special case of GLMs, we model the response variable $y$ as a Bernoulli distribution with parameter $\phi$:
$$
\begin{align*}
p(y \mid x; \theta) &\sim \text{Bernoulli}(\phi) \\
h_{\theta}(x) &= \mathbb{E}[y \mid x; \theta] \\
&= \phi \\
&= \frac{1}{1 + e^{-\eta}} \quad\left(\because \eta = \log\left(\frac{\phi}{1-\phi}\right) \text{ for Bernoulli}\right) \\
&= \frac{1}{1 + e^{-\theta^{T}x}} \blacksquare \\
\end{align*}
$$

### Estimating the Parameters $(\theta)$
We have shown previously that the log likelihood function for the exponential family of distributions is concave in the natural parameter $\eta$. Since $\eta$ is a linear function of $\theta$, the log likelihood function is also concave in $\theta$. This means that we can use a maximum likelihood estimation (MLE) approach to estimate the parameters $\theta$ of a GLM.

We start by writing down the likelihood function for a GLM:
$$
\begin{align*}
L(\theta) &= P(y \mid X; \eta) \\
&= \prod_{i=1}^{m} p(y^{(i)} \mid x^{(i)}; \eta^{(i)}) \quad \left(\text{assuming i.i.d. samples}\right) \\
&= \prod_{i=1}^{m} p(y^{(i)}; \eta^{(i)}) \quad \left(\because y^{(i)} \text{ depends on } x^{(i)} \text{ only through } \eta^{(i)}\right) \\
&= \prod_{i=1}^{m} b(y^{(i)}) \exp(\eta^{(i)T}T(y^{(i)}) - a(\eta^{(i)})) \\
&= \prod_{i=1}^{m} b(y^{(i)}) \exp((\theta^{T}x^{(i)})^{T}T(y^{(i)}) - a(\theta^{T}x^{(i)})) \\
&= \prod_{i=1}^{m} b(y^{(i)}) \exp({x^{(i)}}^T\theta T(y^{(i)}) - a(\theta^{T}x^{(i)})) \\
\end{align*}
$$

Taking the log of the likelihood function, we get the log likelihood function:
$$\begin{align*}
\ell(\theta) &= \sum_{i=1}^{m} \log b(y^{(i)}) + \sum_{i=1}^{m} {x^{(i)}}^T\theta T(y^{(i)}) - \sum_{i=1}^{m} a(\theta^{T}x^{(i)}) \\
&= \sum_{i=1}^{m} \log b(y^{(i)}) + \theta \sum_{i=1}^{m} {x^{(i)}}^{T} T(y^{(i)}) - \sum_{i=1}^{m} a(\theta^{T}x^{(i)}) \\
\end{align*}
$$

To find the maximum likelihood estimator (MLE) for $\theta$, we take the gradient of the log likelihood with respect to $\theta$ and set it to zero:
$$\begin{align*}
\nabla_{\theta} \ell(\theta) &= \sum_{i=1}^{m} 0 + \sum_{i=1}^{m} {x^{(i)}}^T T(y^{(i)}) - \sum_{i=1}^{m} \nabla_{\theta} a(\theta^{T}x^{(i)}) \\
&= \sum_{i=1}^{m} x^{(i)} T(y^{(i)}) - \sum_{i=1}^{m} \frac{\partial a(\theta^{T}x^{(i)})}{\partial \theta} \\
&= \sum_{i=1}^{m} x^{(i)} T(y^{(i)}) - \sum_{i=1}^{m} \frac{\partial a(\eta^{(i)})}{\partial \eta^{(i)}} \frac{\partial \eta^{(i)}}{\partial \theta} \\
&= \sum_{i=1}^{m} x^{(i)} T(y^{(i)}) - \sum_{i=1}^{m} \mathbb{E}[y^{(i)} ; \eta^{(i)}] x^{(i)} \quad \left(\text{by property of exponential family, } \frac{\partial a(\eta)}{\partial \eta} = \mathbb{E}[y ; \eta]\right) \\
&= \sum_{i=1}^{m} x^{(i)} \left(T(y^{(i)}) - \mathbb{E}[y^{(i)} ; \eta^{(i)}]\right) \\
&= \sum_{i=1}^{m} x^{(i)} \left(T(y^{(i)}) - h_{\theta}(x^{(i)})\right) \quad \left(\text{by design choice, } h_{\theta}(x) = \mathbb{E}[y ; \eta]\right) \\
\end{align*}
$$

Thus, the LMS update rule for GLMs becomes:
$$
\begin{align*}
\theta &:= \theta + \alpha \nabla_{\theta} \ell(\theta) &\quad(\text{gradient ascent}) \\
\theta &:= \theta + \alpha \sum_{i=1}^{m} x^{(i)} \left(T(y^{(i)}) - h_{\theta}(x^{(i)})\right) &\quad (\text{batch gradient ascent}) \\
\theta &:= \theta + \alpha x^{(i)} \left(T(y^{(i)}) - h_{\theta}(x^{(i)})\right) &\quad (\text{stochastic gradient ascent}) \\
\end{align*}
$$

In most cases, the sufficient statistic $T(y) = y$, so the update rule simplifies to:
$$
\boxed{\theta := \theta + \alpha x^{(i)} \left(y^{(i)} - h_{\theta}(x^{(i)})\right)}
$$
