# Notations

$$
\begin{align*}

x^{(i)}, (1 \leq i \leq n) &- \text{input variables/features in the $i^{th}$ row of training set} \\
y^{(i)}, (1 \leq i \leq n) &- \text{output/target variable in the $i^{th}$ row of training set} \\
x_{i}, (1 \leq i \leq d) &- i^{\text{th}} \text{ feature column in training set} \\
(x^{(i)}, y^{(i)}) &- \text{training example in $i^{th}$ row of training set} \\
n &- \text{number of training examples} \\
d &- \text{number of input variables/features} \\
\mathcal{X} = \{x: x \in x^{(i)}\} &- \text{Space of input values} \\
\mathcal{Y} = \{y: y \in y^{(i)}\} &- \text{Space of output values} \\
\theta_{i} &- \text{parameters/weights} \\
h: \mathcal{X} \to \mathcal{Y}, h(x) = h_{\theta}(x) &- \text{hypothesis}

\end{align*}
$$

To make it easier to work with some math, we usually define a *dummy feature* $x_0 = 1$ and $\theta_{0}$ is the corresponding *intercept/bias parameter*. The training set or rows of training examples/features $\mathcal{X}$, can then be represented as an $n\times (d + 1)$ matrix. Similarly, the weights/parameters $\mathcal{\theta_{i}}$ and outputs $\mathcal{Y}$ can also be represented as $(d + 1) \times 1 $ and $n \times 1$ vectors respectively:

$$
\begin{align*}

\Chi &= \begin{bmatrix}
    1 & x_{1}^{(1)} & x_{2}^{(1)} & \cdots & x_{d}^{(1)} \\
    1 & x_{1}^{(2)} & x_{2}^{(2)} & \cdots & x_{d}^{(2)} \\
    1 & \vdots & \vdots & \ddots & \vdots \\
    1 & x_{1}^{(n)} & x_{2}^{(n)} & \cdots & x_{d}^{(n)} \\
\end{bmatrix}_{n\times (d+1)} \\

\Theta &= \begin{bmatrix}
    \theta_{0} \\
    \theta_{1} \\
    \theta_{2} \\
    \vdots \\
    \theta_{d} \\
\end{bmatrix}_{(d + 1)\times 1} \\


\mathrm{Y} &= \Chi \times \Theta = \begin{bmatrix}
    y_{1} \\
    y_{2} \\
    \vdots \\
    y_{n} \\
\end{bmatrix}_{n\times 1}

\end{align*}
$$

If $\mathcal{Y}$ is continuous or the space of output variables is infinite, the learning problem is called a **regression problem**. If $\mathcal{Y}$ is discrete or the space of output variables is finite, the learning problem is called a **classification problem**.
